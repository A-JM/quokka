```python
import ray
ray.init()
```

```python
@ray.remote
class Counter(object):
    def __init__(self,c):
        self.value = c

    def increment(self,b):
        self.value += b
        return self.value

    def get_counter(self):
        return self.value

counter_actor = Counter.remote(2)
```

```python
ray.get(counter_actor.increment.remote(2))
```

```python
ray.get(counter_actor.get_counter.remote())
```
